```python
# Conectar Google Drive e configurar ambiente
from google.colab import drive
import os
from pathlib import Path
import glob
from datetime import datetime

drive.mount('/content/drive')

# Caminhos importantes
valetts_path = "/content/drive/MyDrive/ValeTTS"
logs_path = f"{valetts_path}/logs"
reports_path = f"{valetts_path}/reports"
samples_path = f"{valetts_path}/samples"
checkpoints_path = f"{valetts_path}/checkpoints"

print("🔍 Status dos diretórios:")
for name, path in [("Logs", logs_path), ("Reports", reports_path), 
                   ("Samples", samples_path), ("Checkpoints", checkpoints_path)]:
    exists = "✅" if os.path.exists(path) else "❌"
    print(f"   {exists} {name}: {path}")
```

```python
# Iniciar TensorBoard
%load_ext tensorboard
%tensorboard --logdir /content/drive/MyDrive/ValeTTS/logs
```

```python
# Visualizar relatórios LLM mais recentes
import glob
from IPython.display import Markdown, display

def show_latest_reports(limit=3):
    reports = glob.glob(f"{reports_path}/*.md")
    reports.sort(key=os.path.getmtime, reverse=True)
    
    if not reports:
        print("📋 Nenhum relatório LLM encontrado ainda")
        return
    
    print(f"📋 Mostrando {min(limit, len(reports))} relatórios mais recentes:\n")
    
    for i, report in enumerate(reports[:limit]):
        print(f"\n{'='*60}")
        print(f"📄 RELATÓRIO {i+1}: {os.path.basename(report)}")
        print(f"⏰ Modificado: {datetime.fromtimestamp(os.path.getmtime(report))}")
        print(f"{'='*60}")
        
        try:
            with open(report, 'r', encoding='utf-8') as f:
                content = f.read()
                display(Markdown(content))
        except Exception as e:
            print(f"❌ Erro ao ler relatório: {e}")

show_latest_reports()
```

```python
# Verificar checkpoints e amostras
def check_training_progress():
    print("🔍 PROGRESS DO TREINAMENTO\n")
    
    # Checkpoints
    checkpoints = glob.glob(f"{checkpoints_path}/*.ckpt")
    print(f"💾 Checkpoints encontrados: {len(checkpoints)}")
    if checkpoints:
        latest_ckpt = max(checkpoints, key=os.path.getmtime)
        ckpt_time = datetime.fromtimestamp(os.path.getmtime(latest_ckpt))
        ckpt_size = os.path.getsize(latest_ckpt) / 1e6
        print(f"   📁 Mais recente: {os.path.basename(latest_ckpt)}")
        print(f"   ⏰ Modificado: {ckpt_time}")
        print(f"   📊 Tamanho: {ckpt_size:.1f} MB")
    
    print()
    
    # Amostras de áudio
    samples = glob.glob(f"{samples_path}/*.wav")
    print(f"🎵 Amostras de áudio: {len(samples)}")
    if samples:
        latest_sample = max(samples, key=os.path.getmtime)
        sample_time = datetime.fromtimestamp(os.path.getmtime(latest_sample))
        print(f"   📁 Mais recente: {os.path.basename(latest_sample)}")
        print(f"   ⏰ Gerado: {sample_time}")
    
    print()
    
    # Relatórios
    reports = glob.glob(f"{reports_path}/*.md")
    print(f"📋 Relatórios LLM: {len(reports)}")
    if reports:
        latest_report = max(reports, key=os.path.getmtime)
        report_time = datetime.fromtimestamp(os.path.getmtime(latest_report))
        print(f"   📁 Mais recente: {os.path.basename(latest_report)}")
        print(f"   ⏰ Gerado: {report_time}")

check_training_progress()
```

```python
# Reproduzir amostras de áudio mais recentes
import IPython.display as ipd

def play_latest_samples(limit=3):
    samples = glob.glob(f"{samples_path}/*.wav")
    samples.sort(key=os.path.getmtime, reverse=True)
    
    if not samples:
        print("🎵 Nenhuma amostra de áudio encontrada ainda")
        return
    
    print(f"🎵 Reproduzindo {min(limit, len(samples))} amostras mais recentes:\n")
    
    for i, sample in enumerate(samples[:limit]):
        sample_name = os.path.basename(sample)
        sample_time = datetime.fromtimestamp(os.path.getmtime(sample))
        
        print(f"🎤 {i+1}. {sample_name}")
        print(f"⏰ Gerado: {sample_time}")
        
        # Reproduzir áudio
        display(ipd.Audio(sample))
        print()

play_latest_samples()
```
